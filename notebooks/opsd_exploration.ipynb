## 1. Filter out the load_actual_entsoe_transparency fields for each country.

```

wind_onshore_capacity
wind_onshore_generation_actual
wind_onshore_profile
```

There are also timestamp columns: `utc_timestamp`, `cet_cest_timestamp`

As an example case, we'll just filter on the `load_actual_entsoe_transparency`. This is just a placeholder until we know which data is actually useful; it's easy enough to change `base_field_name` to some other column.

````python
base_field_name = 'load_actual_entsoe_transparency' 

data_cols = [k for k in raw_df.columns if base_field_name in k]
col_name_mapping = {k: k.replace(f"_{base_field_name}", "") for k in data_cols}
filter_cols = ['utc_timestamp'] + data_cols
````

````python
raw_df[filter_cols].rename(columns=col_name_mapping).set_index('utc_timestamp')
````
